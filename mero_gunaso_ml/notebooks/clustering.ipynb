```python
from sklearn.cluster import KMeans
import numpy as np

# 6 points — clearly 2 groups
X = np.array([
    [1, 1], [1.5, 1], [1, 1.5],   # group A
    [8, 8], [8.5, 8], [8, 8.5],   # group B
])

km = KMeans(n_clusters=2, random_state=42)
km.fit(X)
print(km.labels_)   # [0 0 0 1 1 1] or [1 1 1 0 0 0]
```
